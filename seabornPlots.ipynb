```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import random
%matplotlib inline
import seaborn as sns

# rescale the data

n0 = sum(1 for line in open('data2000.csv')) - 1
n1 = sum(1 for line in open('data2001.csv')) - 1
n2 = sum(1 for line in open('data2002.csv')) - 1
n3 = sum(1 for line in open('data2003.csv')) - 1
n4 = sum(1 for line in open('data2004.csv')) - 1
n5 = sum(1 for line in open('data2005.csv')) - 1
n6 = sum(1 for line in open('data2006.csv')) - 1
n7 = sum(1 for line in open('data2007.csv')) - 1
n8 = sum(1 for line in open('data2008.csv')) - 1

#Â sample size
s = 20000

skip0 = sorted(random.sample(range(1,n0+1),n0-s))
skip1 = sorted(random.sample(range(1,n1+1),n1-s))
skip2 = sorted(random.sample(range(1,n2+1),n2-s))
skip3 = sorted(random.sample(range(1,n3+1),n3-s))
skip4 = sorted(random.sample(range(1,n4+1),n4-s))
skip5 = sorted(random.sample(range(1,n5+1),n5-s))
skip6 = sorted(random.sample(range(1,n6+1),n6-s))
skip7 = sorted(random.sample(range(1,n7+1),n7-s))
skip8 = sorted(random.sample(range(1,n8+1),n8-s))

data2000 = pd.read_csv('data2000.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip0)
data2001 = pd.read_csv('data2001.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip1)
data2002 = pd.read_csv('data2002.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip2)
data2003 = pd.read_csv('data2003.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip3)
data2004 = pd.read_csv('data2004.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip4)
data2005 = pd.read_csv('data2005.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip5)
data2006 = pd.read_csv('data2006.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip6)
data2007 = pd.read_csv('data2007.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip7)
data2008 = pd.read_csv('data2008.csv', sep=',', names = ['Years','UniqueCarrier', 'DepDelay'], converters ={'Years':int ,'DepDelay':int}, skiprows=skip8)

data = data2000
data = data.append(data2001, ignore_index=True)
data = data.append(data2002, ignore_index=True)
data = data.append(data2003, ignore_index=True)
data = data.append(data2004, ignore_index=True)
data = data.append(data2005, ignore_index=True)
data = data.append(data2006, ignore_index=True)
data = data.append(data2007, ignore_index=True)
data = data.append(data2008, ignore_index=True)
```

```python
sns.set(style="whitegrid", color_codes=True)
plt.figure(figsize=(20, 20))

np.random.seed(sum(map(ord, "categorical")))

sns.stripplot(x="Years", y="DepDelay",hue='UniqueCarrier', data=data, jitter=True);

```

```python
# calculate the current year (or overall) delay probability
def delayProb(df):
    countDelay = 0
    for i in df['DepDelay']:
        if i > 0:
            countDelay += 1
        
    return countDelay/len(df['DepDelay'])
```

```python
dpAll = delayProb(data)
dp2000 = delayProb(data2000)
dp2001 = delayProb(data2001)
dp2002 = delayProb(data2002)
dp2003 = delayProb(data2003)
dp2004 = delayProb(data2004)
dp2005 = delayProb(data2005)
dp2006 = delayProb(data2006)
dp2007 = delayProb(data2007)
dp2008 = delayProb(data2008)

probList = [dp2000, dp2001, dp2002, dp2003, dp2004, dp2005, dp2006, dp2007, dp2008]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008]

plt.figure(figsize=(10, 5))
plt.ylim(0,1)
plt.plot(years, probList, marker='o')

print("delay probability for all  is", dpAll)
print("delay probability for 2000 is", dp2000)
print("delay probability for 2001 is", dp2001)
print("delay probability for 2002 is", dp2002)
print("delay probability for 2003 is", dp2003)
print("delay probability for 2004 is", dp2004)
print("delay probability for 2005 is", dp2005)
print("delay probability for 2006 is", dp2006)
print("delay probability for 2007 is", dp2007)
print("delay probability for 2008 is", dp2008)
```
