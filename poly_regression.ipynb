```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
airlines = pd.read_csv("airlines.csv")
airline = airlines.loc[:,"netIncome":"fleetSize"]
```

```python
import seaborn
plt.figure(figsize=(12,8))
scat1=seaborn.regplot(x="netIncome", y="avarageDelay", scatter=True, order=2, data=airline)
plt.ylim(-2.5,15)
plt.xlabel('net income')
plt.ylabel('average delay')
```

```python
plt.figure(figsize=(12,8))
scat2=seaborn.regplot(x="market", y="avarageDelay", scatter=True, order=2, data=airline)
plt.ylim(-2.5,15)
plt.xlabel('market Share')
plt.ylabel('average delay')
```

```python
plt.figure(figsize=(12,8))
scat3=seaborn.regplot(x="employee", y="avarageDelay", scatter=True, order=2, data=airline)
plt.ylim(-2.5,15)
plt.xlabel('employee number')
plt.ylabel('average delay')
```

```python
plt.figure(figsize=(12,8))
scat4=seaborn.regplot(x="fleetSize", y="avarageDelay", scatter=True, order=2, data=airline)
plt.ylim(-2.5,15)
plt.xlabel('fleet size')
plt.ylabel('average delay')
```
